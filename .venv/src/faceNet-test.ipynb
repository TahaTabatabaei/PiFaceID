```python
# Step 1: Load the TensorFlow v1 model from the .pb file
def load_tf_v1_model(pb_model_path):
    with tf.compat.v1.Session() as sess:
        with tf.io.gfile.GFile(pb_model_path, 'rb') as f:
            graph_def = tf.compat.v1.GraphDef()
            graph_def.ParseFromString(f.read())
            tf.import_graph_def(graph_def, name='')
    return sess

# Replace 'model.pb' with the path to your .pb model file
pb_model_path = model_path
sess = load_tf_v1_model(pb_model_path)

inpute_tensor = tf.compat.v1.get_default_graph().get_tensor_by_name('input:0')
output_tensor = tf.compat.v1.get_default_graph().get_tensor_by_name("embeddings:0")
```

```python


# Step 2: Convert the TensorFlow v1 model to TensorFlow Lite
def convert_to_tflite(sess, tflite_model_path):
    converter = tf.compat.v1.lite.TFLiteConverter.from_session(sess, input_tensors=[inpute_tensor], output_tensors=[output_tensor])
    converter.allow_custom_ops = True

    
    tflite_model = converter.convert()
    with open(tflite_model_path, 'wb') as f:
        f.write(tflite_model)
    print(f"TensorFlow Lite model saved to: {tflite_model_path}")

# Replace 'converted_model.tflite' with the desired TFLite model file path
# tflite_model_path = 'converted_model.tflite'
convert_to_tflite(sess, tflite_model_path)
```

```python
# Instantiate an interpreter for each model
tf_model = tf.lite.Interpreter(tflite_model_path)

# Allocate memory for each model
tf_model.allocate_tensors()
```
